```python
import duckdb
```

```python
from pathlib import Path
PROJECT_ROOT = Path.cwd().parents[0]  # adjust if you launch Jupyter from root
DB_PATH = PROJECT_ROOT / "data" / "market_data.duckdb"

con = duckdb.connect(str(DB_PATH))
```

```python
con.execute("SELECT * FROM pipeline_metrics ORDER BY run_started_at DESC").fetchdf()
```

```python
con.execute("""
    SELECT
        run_started_at,
        messages_processed,
        elapsed_seconds,
        messages_processed / NULLIF(elapsed_seconds, 0) AS msgs_per_sec
    FROM pipeline_metrics
    ORDER BY run_started_at
""").fetchdf()
```

```python
con.close()
```
